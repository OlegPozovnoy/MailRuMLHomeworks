```
# без ограничения общности это уже зашифрованное сообщение

text = "Реализуйте базовый частотный метод по Шерлоку Холмсу:\
подсчитайте частоты букв по корпусам (пунктуацию и капитализацию можно просто опустить, а вот пробелы лучше оставить);\
возьмите какие-нибудь тестовые тексты (нужно взять по меньшей мере 2-3 предложения, иначе совсем вряд ли сработает), зашифруйте их посредством случайной перестановки символов;\
расшифруйте их таким частотным методом.\
"
text = preprocess_text(text)
data = preprocess_text(data)

def create_ngram(text, n):
  ngrams = {}
  text = text.lower()
  for i in range(len(text)-n+1):
    if text[i:i+n] not in ngrams:
      ngrams[text[i:i+n]] = 0
    ngrams[text[i:i+n]] = ngrams[text[i:i+n]]+1
  return ngrams


def map_dict_by_freq(target, source):
  source_counts = [(k,v) for k,v in sorted(source.items(), key=lambda item: item[1], reverse=True)]
  target_counts = [(k,v) for k,v in sorted(target.items(), key=lambda item: item[1], reverse=True)]
  codes = {}
  for i in range(len(target_counts)):
    codes[target_counts[i][0]] = source_counts[i][0]
  #print(codes)
  return codes
  
def encode_text(text, decoder_dict, n):
  result = []
  for i in range(int(len(text)/n)):
    result.append(decoder_dict[text[i*n: (i+1)*n]]) 
  return result

def full_cycle(source_text, target_text, n):
  source_dict = create_ngram(source_text, n)  
  print("Source",source_dict)
  target_dict = create_ngram(target_text, n)
  print("Dest",target_dict)
  translate_dict =  map_dict_by_freq(target_dict, source_dict)
  print("Translation:", translate_dict)
  return encode_text(target_text, translate_dict, n)


print("".join(full_cycle(data, text, 1)))
print(text)
```

```
import random

def pick_char_from_dist(dist):
  seed = random.uniform(0, 1) * dist[0]
  for v in dist[1]:
    seed -= v[1]
    if (seed<=0):
      return v[0]
  return v[0]
```

```
from random import randrange
from tqdm import tqdm
from array import array
from functools import reduce

# расшифруем строку по словарю
def get_current_string(coded, dict):
  return reduce(lambda x,y: x+y, map(lambda x: dict[x], coded), "")
  #return "".join(map(lambda x: dict[x], coded))


def create_initial_state(dict1, dict2):
  result = {} 
  while len(dict1) > 0:
    total_count = 0
    next = []
    for k1,v1 in dict1.items():
      for k2,v2 in dict2.items():
        total_count += v1*v2
        next.append((k1+k2, v1*v2))
    #print((total_count,next))
    pick = pick_char_from_dist((total_count,next))
    result[pick[0]] = pick[1]
    del dict1[pick[0]]
    del dict2[pick[1]]  
  return result

# Используем алгоритм Гиббса
def translate(coded, data, niter = 1000000, initial_dict = None):
  print("длина сообщения", len(coded)) 
  chars_coded = set(coded)
  characters = set(data)
  print(characters)
  print(create_ngram(data, 1))
  
  if initial_dict == None:
    initial_dict = create_initial_state(create_ngram(coded, 1),create_ngram(data, 1))
  #initial_dict = map_dict_by_freq(create_ngram(coded, 1),create_ngram(data, 1))
  print('Первоначальный словарь', initial_dict)

  # создадим для каждого символа массив индексов куда он входит
  positions_dict = {}
  for s in set(coded):
    positions_dict[s] = [i for i in range(len(coded)) if coded[i] == s ]

  print(positions_dict)
  # напишем наивный перевод
  print("Наивный перевод")
  print(get_current_string(coded, initial_dict))
  print(coded)

  # частотный словарь биграм - фиксируем перевод каждого символа кроме одного, а перевод одного семплируем
  data_dict = create_ngram(data, 2)
  print([(k,v) for k,v in sorted(data_dict.items(), key=lambda item: item[1], reverse=True)])

  for k in tqdm(range(niter)):
    if k%100000 == 0 and k>0:
      print(k)
      print((get_current_string(coded, initial_dict)))
      print(coded)
      print(initial_dict)
    for s in chars_coded: # source char
      full_count = 0
      freq_list = []
      for m in characters: # dest char
        if m == initial_dict[s] or list(initial_dict.values()).count(m) == 0:
          char_count = 1
          for pos in positions_dict[s]:
            left = coded[pos-1] if pos > 0 else None
            right = coded[pos+1] if pos < len(coded) -1 else None
            if pos==0: # везде 1 по дефолту т.к. 
              char_count *= data_dict.get(' '+ m, 1) * data_dict.get(m + initial_dict[right], 1)/10.
            elif pos == len(coded) - 1: 
              char_count *= data_dict.get(initial_dict[left] + m, 1) * data_dict.get(m + ' ', 1)/10.
            else:
              char_count *= data_dict.get(initial_dict[left] + m, 1) * data_dict.get(m + initial_dict[right], 1)/10.
            if char_count == 0:
              break
          full_count += char_count
          if char_count > 0:
            freq_list.append((m, char_count))
      if (len(freq_list)>0): 
        initial_dict[s] = pick_char_from_dist((full_count,freq_list))
  print("Результат:")
  print("Финальный словарь:", initial_dict)
  print((get_current_string(coded, initial_dict)))
  print(coded)
  return get_current_string(coded, initial_dict), initial_dict

```

Попробуем алгоритм Метрополиса - Гастингса, выберем каждый раз n переменных словаря, посчитаем провдоподобие строки, и будем переходить или нет, n будем тюнить, чем выше n тем выше дисперсия, так что будем ее постепенно понижать, судя по Гиббсу n=1 мало

```
from functools import reduce

def metropolis_step(text, current_dict, n, full_charset_from, full_charset_to, bigrams):
  
  vars_to_change = random.sample(list(full_charset_from), n)

  #print('vars_to_change', vars_to_change)

  new_dict = current_dict.copy()
  
  remaining_chars = full_charset_to.copy()
  for k, v in current_dict.items():
    if k not in vars_to_change:
      remaining_chars.remove(current_dict[k])
  
  #print('remaining_chars', len(remaining_chars))

  new_dict_values = random.sample(list(remaining_chars), n)
  #print('new_dict_values', new_dict_values)
  
  for i in range(n):
    new_dict[vars_to_change[i]] = new_dict_values[i]

  # посчитаем правдоподобие на новых словарях
  str1 = ' ' + get_current_string(text, new_dict) + ' '
  str2 = ' ' + get_current_string(text, current_dict) + ' '

  #print(str1)
  #print(str2)

  log_p1 = reduce(lambda x,y: x+y,map(lambda x: math.log(bigrams.get(str1[x:x+2],1)),range(len(str1)-1)))
  log_p2 = reduce(lambda x,y: x+y,map(lambda x: math.log(bigrams.get(str2[x:x+2],1)),range(len(str1)-1)))

  p = math.exp(log_p1 - log_p2)

  # просемплируем, перейдем в новую точку или нет
  if random.uniform(0, 1) < p:
    return new_dict
  else:
    return current_dict


def translate_new(coded, data, niter = 1000000, initial_dict = None, n=6):
  chars_coded = set(coded)
  characters = set(data)
  
  if initial_dict == None:
    initial_dict = create_initial_state(create_ngram(coded, 1),create_ngram(data, 1))
  print('Первоначальный словарь', initial_dict)

  print("Наивный перевод")
  print(get_current_string(coded, initial_dict))
  print(coded)

  data_dict = create_ngram(data, 2)
  print([(k,v) for k,v in sorted(data_dict.items(), key=lambda item: item[1], reverse=True)])

  for k in tqdm(range(niter)):
    if k%100000 == 0 and k>0:
      print(k)
      print((get_current_string(coded, initial_dict)))
      print(coded)
      print(initial_dict)
      print('n=', (niter - k)**2 * (len(chars_coded)-1)/niter**2 + 1)
      current_str = ' ' + get_current_string(coded, initial_dict) + ' '
      print(reduce(lambda x,y: x+y,map(lambda x: math.log(data_dict.get(current_str[x:x+2],1)),range(len(current_str)-1))))

    n = int((niter - k)**2 * (len(chars_coded)-1)/niter**2 ) + 1
    initial_dict = metropolis_step(coded, initial_dict, n, chars_coded, characters, data_dict)


  print("Результат:")
  print("Финальный словарь:", initial_dict)
  print((get_current_string(coded, initial_dict)))
  print(coded)
  return get_current_string(coded, initial_dict), initial_dict
```
